#### ¿Cuales son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
#pero nosotros queremos el valro de los alquileres
datos['Valor'].mean()
#lo que hace es sumar todos los datos y dividirlos por el largo de esto obteniendo el promedio del conjunto
#no lo qu ebuscamos que es el promedio por tipo de propiedad
```

```python
#para eso necesitamos agrupar los datos por el tipo de datos y asi calcular el promedio
#utilizamos la funcion groupby, pero realmente nos interesa el valor, para que no nos salgass el
#warning, especificamos el numero_only=true, especificando solamente numeros
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
#necesitamos especificar la columna dado que necesitamos una serie de datos especifica
#deriubado que lo anterior esta muy diluido
#por lo tanto agregaremos valor dadoq ue ahi estan los valores monetarios de los tipos
datos.groupby('Tipo')['Valor'].mean()
#nos da los valores promedio de cada uno de los tipos de inmuebles
```

```python
#ahora para una mejor vizualisacion lo colocamos en un formato de data frame
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
#nos da en orden acendente
```

```python
df_tipo_precio=datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
#lo ggraficaremos con una funcion de pantas que es .plot
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='blue')
#nos devuelve que la mayuoria de los edificios en alquiler son industriaes y edificios completos
```

```python
#nos digeron que solo se enfocara en los muebles residenciales, dado que tenemos que segregar entre muebles comerciales e industriales
#una forma de hacerlo no solo con corchete y el nombre del artributo, si no colocar el punto y el nombre del atributo
datos.Tipo.unique()
#tenemoss una lista de los atributos de nuestra base
```

###Remover registros inconsistentes

Hay varias formas de crear columnas en un DataFrame en Pandas. Algunos ejemplos son:

1 - Asignación directa de valores a una nueva columna:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
Copia el código
2 - Creación de una columna a partir de operaciones entre otras columnas:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
Copia el código
3 - Uso del método assign() para crear una nueva columna:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
Copia el código
4 - Uso del método apply() para aplicar una función a una columna existente y crear una nueva columna:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)

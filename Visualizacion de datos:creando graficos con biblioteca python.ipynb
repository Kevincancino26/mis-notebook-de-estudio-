## Creando subplots en unas direcciones

pero tenemos la posibilidad de teer mas informacion estadistica de la migracion de canda con otro grafico como el grafico de bigotes o el grafico de caja, El gráfico de caja o box plot es una herramienta estadística que nos permite visualizar la distribución de un conjunto de datos. Este tipo de gráfico muestra de forma clara cinco valores clave: el mínimo, el primer cuartil (Q1), la mediana (Q2), el tercer cuartil (Q3) y el máximo. Además, nos ayuda a identificar posibles valores atípicos (outliers).

Gracias a esta representación, podemos entender rápidamente cómo se comportan los datos en un área determinada, si están más concentrados en un rango, si hay asimetría, o si existen valores extremos. También es útil para comparar distribuciones entre diferentes grupos o categorías

para eso vamos hacer nuestro ejercicio desde el inicio

```python
fig,axs=plt.subplots(1,2,figsize=(10,3))#el uno y el dos representan que estamos creando dos graficas asi especificamos el numero de graficas
#como trabajamos con dos figuras podemos usar una variable axs
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigracion de colombianos hacia canada \n 1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
#aqui hacemos nuestro grafico de cajas donde al poner los numeros dentro de los corchetes indicamos cual grafica va cada dato a la 1 o la dos
axs[1].boxplot(datos_col['Inmigrantes'])# como ablamos de un box plox, este unicamente medira una variable, solo mediremos la cantidad de migrantes colombianos
axs[1].set_title('Boxplot de la inmigracion \n 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de inmigrantes')
axs[1].grid()


plt.show()
```

###creando sybplots en dos direcciones

podemos crear ua o mas direcciones entre las figuras co subplots es posible, ahora lo harmos con otros paises , argentina, brasil y peri

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
#ajustar los subplots
fig.subplots_adjust(hspace=0.5,wspace=0.3)#h = altura high y w = weit ancho
#titulo para todo el lienzo
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos \nhacia Canada de 1980 hasta 2013')

#CREANDO NUESTROS GRFICOS
axs[0,0].plot(d_f.loc['Colombia',años])
axs[0,0].set_title('Colombia')

axs[0,1].plot(d_f.loc['Brasil',años])
axs[0,1].set_title('Brasil')

axs[1,0].plot(d_f.loc['Argentina',años])
axs[1,0].set_title('Argentina')


axs[1,1].plot(d_f.loc['Perú',años])
axs[1,1].set_title('Perú')
#estableciendo un valro minimo para que todas esten en la misma escala
ymin=0
ymax=7000
#usamos el metodo ravel por que nos permite tomar un array unideccional y ponerlo en una miltidireccional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


#como las graficas tienen los mismo ejes y los mismo años podemos resolverlo
#utilizando un laso for
for ax in axs.flat:#nos metemos dentro de la propiedad axs.flag que nos permite iterar y hacer las isgientes operaciones
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Años')
  ax.set_ylabel('Numero de migrantes')
  ax.grid()
plt.show()
```

 ### Buenas practicas de creacion de los subplots

Usar títulos claros y concisos: el título de cada subplot debe ser corto y descriptivo para que el lector pueda rápidamente entender lo que se está mostrando en cada elemento. Además, si estás comparando diferentes conjuntos de datos, puede ser útil añadir un subtítulo explicando lo que se está comparando.

Mantener la misma escala en los ejes: para evitar distorsiones en la comparación entre los subplots, es importante mantener la misma escala en los ejes X e Y en todos ellos. Esto se puede hacer utilizando las funciones set_xlim() y set_ylim().

Evitar la superposición de gráficos: es importante asegurarse de que cada subplot esté claramente separado de los demás, sin superposición. Si hay una superposición, la figura puede volverse confusa y difícil de entender. Una manera de evitar superposiciones es ajustar el tamaño de los subplots para que haya espacio suficiente entre ellos. Además, podemos añadir la función fig.subplots_adjust() que recibe el parámetro hspace. Controla el espaciado vertical entre los subplots y el parámetro wspace, que controla el espaciado horizontal. El valor por defecto de estos parámetros es 0.2, pero puedes ajustarlos según tus necesidades.

Recuerda que el valor pasado a estos parámetros es un número decimal que representa la fracción del tamaño de la figura, por ejemplo, 0.5 significa que el espaciado será del 50% de la altura/anchura de la figura.

```python
fig , ax = plt.subplots(figsize=(10,5))
#cambiando el espesor de la linea con linewit o lw: esto es la cantidad de pizeles a utilizar y podemos coloar marcadores
ax.plot(datos_col['Año'],datos_col['Inmigrantes'],lw=2,marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))#lo que hacemos es utilizar un localizador multiple donde le indicamos que de 5 en 5 vaya colocando mis ticsk en automatico
ax.set_title('Inmigracion de colombianos hacia Canada \n1980-2023',fontsize=18, loc='left')
#localizamos el titulo hacia la izauierda con el parametro loc
#usamos el comando para introducir el tirulo en el eje x
ax.set_xlabel('Año',fontsize=14)
ax.set_ylabel('Numero de inmigrantes',fontsize=14)
ax.yaxis.set_tick_params(labelsize=12) #de esta manera cambaimos el tamaño del numero
ax.xaxis.set_tick_params(labelsize=12)
plt.grid()
plt.show()
```

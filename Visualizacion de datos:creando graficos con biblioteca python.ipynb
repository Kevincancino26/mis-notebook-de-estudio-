Con el uso del plano cartesiano, constituido por el eje X y el eje Y, es posible crear un sistema de coordenadas bidimensional que facilita la localización y análisis de datos. Estos ejes también son conocidos por los siguientes nombres:

* eje x: eje de las abscisas.
* eje y: eje de las ordenadas.

Los ejes son componentes esenciales debido a sus combinaciones, que posibilitan una mayor precisión en la identificación de puntos en un espacio bidimensional. * El eje X es horizontal.
* El eje Y es vertical.
* En el centro tenemos el origen, el valor 0 (cero).
* Hacia la derecha y hacia arriba tenemos valores positivos.
* Hacia la izquierda y hacia abajo tenemos valores negativos.

##Creando el primer grafico con matplotlib

ahora que tenemos nuestro primer data frame, crearemos una grafica que nos permita analisar el patron de tendencia de los migrantes colombianos

```python
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
```

nos muestra una grafica con los datos sobresalidos y tambien nos muestra en donde fue asignada en el espacio en memora

ahora vamos a mejorar el grafico

```python
plt.figure(figsize=(10,5))#esto determina el tañaño de la figura, los numeros son en pulgados
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])#esto se refiere a cada uno de las rallas del eje x que señalan los titulos de los elementos en este caso los años
#la idea es solo visualizar alguno de los 30 años que tenemos solo un pequeño lapso de dats
#asi pordemos eliminar el mensaje de donde se almaceno
plt.show()
```

ahora nuestro grafico es mas visible y vemos que tenemos un intervalo de años en el grafico x de 5 en 5, asi podemos hacerlo mismo con y usando ystiks para ponerlo en intervalos, junto con el ajsute del tamaño de nuestra figura

## Añadiendo titulos

resumiendo lo anterior: hicimos un grafico especificando el tamaño, asi como creamos un grafico de linea con la funcion plot, donde x es años y "y" es la cantidad de migrantes y agregamos ticks para que no se viera ensimado

ahora continuaremos con el analisis

```python
plt.figure(figsize=(10,5))#esto determina el tañaño de la figura, los numeros son en pulgados
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])#esto se refiere a cada uno de las rallas del eje x que señalan los titulos de los elementos en este caso los años
#vamos a ponerle titulo
plt.title('Inmigracion de colombianos hacia Canada')
#etiqueta para eje x y y
plt.xlabel('Años')
plt.ylabel('Num.Inmigrantes')
plt.show()#es importante que este sea el ultimo comando, dado que si colocamos algo despues de este, no aparecera, no se ejecutara
```

vamos a crear una variable para almacenar nuestra grafica donde la variable sera fig el espacio en la grafica y la grafica ax

usando el metodo subplots podemos crear varias graficas al mismo tiempo 2,3,4

```python
fig , ax = plt.subplots(figsize=(10,5))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.show()
```

ahora atraves de una funcion podemos modificar o alterar nuestra grafica

```python
fig , ax = plt.subplots(figsize=(10,5))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))#lo que hacemos es utilizar un localizador multiple donde le indicamos que de 5 en 5 vaya colocando mis ticsk en automatico
plt.show()
```
